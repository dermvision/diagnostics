###Clean Taxonomy

```python
path_filter = lambda s: all(c not in uppercase for c in s)
P = read_paths('paths_messy.txt', path_filter)
```

###Save Taxonomy

```python
T = taxonomy_from_paths(P)
capitalize_leaves(T)
save_taxonomy(T, 'taxonomy')
P = paths_from_taxonomy(T)
save_paths(P, 'paths_full.txt')
remove_leaves(T)
P = paths_from_taxonomy(T)
save_paths(P, 'paths.txt')
```

###Sanity Check

```python
P = read_paths('paths.txt')
P == paths_from_taxonomy(taxonomy_from_paths(P))
```
